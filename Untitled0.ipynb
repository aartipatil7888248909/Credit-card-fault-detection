```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets"
r = requests.get(url)
print(r)
```

```python
soup = BeautifulSoup(r.text,"lxml")
soup
```

```python
names = soup.find_all("a",class_="title")
names
```

```python
product_name = []
for i in names:
    name = i.text
    product_name.append(name)
print(product_name)
```

```python
product_name = [product.strip().replace('\n','').replace('\t','') for product in product_name]
product_name
```

```python
prices = soup.find_all("h4",class_="price float-end card-title pull-right")
print(prices)
```

```python
prices_list = []
for i in prices:
    price = i.text
    prices_list.append(price)
print(prices_list)
```

```python
prices_list[:] = [price.strip().replace('\n','\t').replace('\n','\t') for  price in prices_list]
prices_list
```

```python
desc = soup.find_all("p",class_="description")
desc
```

```python
desc_list = []
for i in desc:
    desc = i.text
    desc_list.append(desc)
print(desc_list)
```

```python
desc_list[:] = [description.strip().replace('\n', '').replace('\t', '') for description in desc_list]
desc_list
```

```python
reviews = soup.find_all("p",class_="review-count float-end")
reviews
```

```python
reviews_list = []
for i in reviews:
    review = i.text
    reviews_list.append(review)
print(reviews_list)
```

```python
reviews_list = [review.strip().replace('\n', '').replace('\t', '') for review in reviews_list]
reviews_list
```

```python
df = pd.DataFrame({"Product name":product_name,"Prices":prices_list,"Description":desc_list,"Reviews":reviews_list})
df
```

```python
reviews_list = [review.strip().replace('\n', '').replace('\t', '') for review in reviews_list]
reviews_list
```
